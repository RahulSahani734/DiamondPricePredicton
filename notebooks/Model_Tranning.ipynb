```python
## data integrition step
data = pd.read_csv("D:/Dimond_prise_prediction_project/notebooks/data/gemstone.csv")
data.head() 
```

```python
data.drop(data.columns[0],axis = 1,inplace=True)
data.head()
```

```python
data['Price'] = data['price']
data.drop('price',axis = 1,inplace = True)
data
```

```python
data = data.astype({'carat':'float', 'depth':'float', 'table':'float', 'x':'float', 'y':'float', 'z':'float', 'Price':'int'})
```

```python
# Independent and Dpendent Feature
X = data.drop('Price',axis = 1)
Y = data[['Price']]
```

```python
X,Y,data
```

```python
# segregate numerical and categorical columns
categorical_columns = X.columns[X.dtypes == 'object']
numerical_columns = X.columns[X.dtypes != 'object']
print("numerical columns",numerical_columns)
print("categoricall columns",categorical_columns)
```

```python
from sklearn.impute import SimpleImputer ## Handlilng missing values
from sklearn.preprocessing import StandardScaler # Handling Feature scaling
from sklearn.preprocessing import OrdinalEncoder ## Orninal Encoding
## Pipeling
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.30,random_state = 30)
print("x_train columns:", x_train.columns)
print("x_test columns:", x_test.columns)
print("y_test columns:", y_test.columns)
print("y_test columns:", y_test.columns)
```

```python
# Numerical pipeline
num_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)



# Categorical pipeline 

cut_categories = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

cate_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinal encoding',OrdinalEncoder(categories =[cut_categories,color_categories,clarity_categories])),
        ('scaler',StandardScaler())
    ]
)

 # column transformer
preprocessor = ColumnTransformer([
    ('numerical_pipeline',num_pipeline,numerical_columns),
    ('categorical_pipeline',cate_pipeline,categorical_columns)
])
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(x_train),columns=preprocessor.get_feature_names_out())
X_test= pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mse)
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square

# Train multiple models and evaluate
models = {
    'linearregression': LinearRegression(),
    'lasso': Lasso(),
    'ridge': Ridge(),
    'elasticnet': ElasticNet()
}

trained_model_list = []
model_list = []
r2_list = []

for model_name, model in models.items():
    model.fit(X_train, y_train)
    
    # Make prediction
    y_pred = model.predict(X_test)
    
    # Evaluate model
    mae, rmse, r2_square = evaluate_model(y_test, y_pred)
    
    print(f"{model_name} model training performance:")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2 Score: {r2_square * 100:.2f}%")
    
    # Append results to lists
    model_list.append(model_name)
    trained_model_list.append(model)
    r2_list.append(r2_square)
    print("___---___---___---___---___---___---___---___---___---___---",end ="\n")
```
